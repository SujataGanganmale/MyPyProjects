```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic
```

```python
titanic.head()
```

```python
sns.jointplot(x='fare',y='age', data=titanic)
```

```python
sns.displot(titanic['fare'],bins=30)
plt.xlim(0,700)
plt.ylim(0,500)
```

```python
sns.boxplot(x='class',y='age', data=titanic)
```

```python
sns.swarmplot(x='class',y='age', data=titanic,size=3)
```

```python
sns.catplot(x='sex',data=titanic,kind='count')
```

```python
sns.heatmap(titanic.corr(),cmap='Set1')
```

```python
g=sns.FacetGrid(titanic,col='sex')
g.map(plt.hist,"age").add_legend()
```
