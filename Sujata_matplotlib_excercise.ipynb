```python
import numpy as np
x=np.arange(0,100)
y=x*2
z=x**2
```

```python
import matplotlib.pyplot as plt
```

## Exercise 1

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

## Excercise 2

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,.2,.2])
ax1.plot(x,y)
ax2.plot(x,y)
```

## Exercise 3

```python
fig=plt.figure()
ax3=fig.add_axes([0,0,1,1])
ax4=fig.add_axes([0.2,0.5,0.4,0.4])
ax4.set_xlim([20,22])
ax4.set_ylim([30,50])
ax4.plot(x,y)
ax3.plot(x,z)
```

## Exercise 4

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(x,z)
```
